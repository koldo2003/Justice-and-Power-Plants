```python
import pandas as pd
```

```python
percentile = pd.read_csv("percentileCENSUSdata.csv")
population = pd.read_csv("cleanedCENSUSdata.csv")
```

```python
ethnicity_perc = percentile[['STNAME','CTYNAME', 'TOT_POP','NHWA_MALE','NHWA_FEMALE','NHBA_MALE','NHBA_FEMALE','NHIA_MALE','NHIA_FEMALE','NHAA_MALE','NHAA_FEMALE','NHNA_MALE','NHNA_FEMALE','H_MALE','H_FEMALE']]
```

```python
ethnicity_perc.head()
```

```python
ethnicity = population[['STNAME','CTYNAME', 'TOT_POP','NHWA_MALE','NHWA_FEMALE','NHBA_MALE','NHBA_FEMALE','NHIA_MALE','NHIA_FEMALE','NHAA_MALE','NHAA_FEMALE','NHNA_MALE','NHNA_FEMALE','H_MALE','H_FEMALE']]
```

```python
ethnicity.head()
```

```python
sum_column = ethnicity_perc['NHWA_MALE']+ ethnicity_perc['NHWA_FEMALE']
ethnicity_perc['NHWA_perc'] = sum_column

sum_column = ethnicity_perc['NHBA_MALE']+ ethnicity_perc['NHBA_FEMALE']
ethnicity_perc['NHBA_perc'] = sum_column

sum_column = ethnicity_perc['NHIA_MALE']+ ethnicity_perc['NHIA_FEMALE']
ethnicity_perc['NHIA_perc'] = sum_column

sum_column = ethnicity_perc['NHAA_MALE']+ ethnicity_perc['NHAA_FEMALE']
ethnicity_perc['NHAA_perc'] = sum_column

sum_column = ethnicity_perc['NHNA_MALE']+ ethnicity_perc['NHNA_FEMALE']
ethnicity_perc['NHNA_perc'] = sum_column

sum_column = ethnicity_perc['H_MALE']+ ethnicity_perc['H_FEMALE']
ethnicity_perc['H_perc'] = sum_column

ethnicity_perc = ethnicity_perc[['STNAME','CTYNAME', 'NHWA_perc', 'NHBA_perc','NHIA_perc','NHAA_perc','NHNA_perc','H_perc']]
ethnicity_perc.head()
```

```python
sum_column = ethnicity['NHWA_MALE']+ ethnicity['NHWA_FEMALE']
ethnicity['NHWA'] = sum_column

sum_column = ethnicity['NHBA_MALE']+ ethnicity['NHBA_FEMALE']
ethnicity['NHBA'] = sum_column

sum_column = ethnicity['NHIA_MALE']+ ethnicity['NHIA_FEMALE']
ethnicity['NHIA'] = sum_column

sum_column = ethnicity['NHAA_MALE']+ ethnicity['NHAA_FEMALE']
ethnicity['NHAA'] = sum_column

sum_column = ethnicity['NHNA_MALE']+ ethnicity['NHNA_FEMALE']
ethnicity['NHNA'] = sum_column

sum_column = ethnicity['H_MALE']+ ethnicity['H_FEMALE']
ethnicity['H'] = sum_column

ethnicity = ethnicity[['STNAME','CTYNAME', 'NHWA', 'NHBA','NHIA','NHAA','NHNA','H']]
ethnicity.head()
```

```python
ethniccensus = ethnicity.merge(ethnicity_perc)
```
