```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
EIA = pd.read_csv("cleanedEIAdata.csv")
percentile = pd.read_csv("percentileCENSUSdata.csv")
population = pd.read_csv("cleanedCENSUSdata.csv")
```

```python
percentile.head()
```

```python
WhiteVPOC_percentile = percentile[['STNAME','CTYNAME', 'TOT_POP','WA_MALE','WA_FEMALE']]
WhiteVPOC_total = population[['STNAME','CTYNAME', 'TOT_POP','WA_MALE','WA_FEMALE']]
```

```python
sum_column = WhiteVPOC_percentile['WA_MALE']+ WhiteVPOC_percentile['WA_FEMALE']
WhiteVPOC_percentile['WA'] = sum_column

sum_column = WhiteVPOC_percentile['TOT_POP'] - WhiteVPOC_percentile['WA']
WhiteVPOC_percentile['POC'] = sum_column

WhiteVPOC_percentile = WhiteVPOC_percentile[['STNAME','CTYNAME', 'TOT_POP', 'WA','POC']]
WhiteVPOC_percentile.head()
```

```python
sum_column = WhiteVPOC_total['WA_MALE']+ WhiteVPOC_total['WA_FEMALE']
WhiteVPOC_total['WA'] = sum_column

sum_column = WhiteVPOC_total['TOT_POP'] - WhiteVPOC_total['WA']
WhiteVPOC_total['POC'] = sum_column

WhiteVPOC_total = WhiteVPOC_total[['STNAME','CTYNAME', 'TOT_POP', 'WA','POC']]
WhiteVPOC_total.head()
```

```python
WhiteVBlackVHispanic_percentile = percentile[['STNAME','CTYNAME', 'TOT_POP','NHWA_MALE','NHWA_FEMALE','NHBA_MALE','NHBA_FEMALE', 'H_MALE','H_FEMALE']]
WhiteVBlackVHispanic_total = population[['STNAME','CTYNAME', 'TOT_POP','NHWA_MALE','NHWA_FEMALE','NHBA_MALE','NHBA_FEMALE', 'H_MALE','H_FEMALE']]
WhiteVBlackVHispanic_percentile.head()
```

```python
sum_column = WhiteVBlackVHispanic_percentile['NHWA_MALE']+ WhiteVBlackVHispanic_percentile['NHWA_FEMALE']
WhiteVBlackVHispanic_percentile['NHWA'] = sum_column

sum_column = WhiteVBlackVHispanic_percentile['NHBA_MALE']+ WhiteVBlackVHispanic_percentile['NHBA_FEMALE']
WhiteVBlackVHispanic_percentile['NHBA'] = sum_column

sum_column = WhiteVBlackVHispanic_percentile['H_MALE']+ WhiteVBlackVHispanic_percentile['H_FEMALE']
WhiteVBlackVHispanic_percentile['H'] = sum_column

WhiteVBlackVHispanic_percentile = WhiteVBlackVHispanic_percentile[['STNAME','CTYNAME', 'TOT_POP', 'NHWA','NHBA','H']]
WhiteVBlackVHispanic_percentile.head()
```

```python
sum_column = WhiteVBlackVHispanic_total['NHWA_MALE']+ WhiteVBlackVHispanic_total['NHWA_FEMALE']
WhiteVBlackVHispanic_total['NHWA'] = sum_column

sum_column = WhiteVBlackVHispanic_total['NHBA_MALE']+ WhiteVBlackVHispanic_total['NHBA_FEMALE']
WhiteVBlackVHispanic_total['NHBA'] = sum_column

sum_column = WhiteVBlackVHispanic_total['H_MALE']+ WhiteVBlackVHispanic_total['H_FEMALE']
WhiteVBlackVHispanic_total['H'] = sum_column

WhiteVBlackVHispanic_total = WhiteVBlackVHispanic_total[['STNAME','CTYNAME', 'TOT_POP', 'NHWA','NHBA','H']]
WhiteVBlackVHispanic_total.head()
```

```python
WhiteHispanicVPOC_percentile = percentile[['STNAME','CTYNAME', 'TOT_POP','NHWA_MALE','NHWA_FEMALE','H_MALE','H_FEMALE']]
WhiteHispanicVPOC_total = population[['STNAME','CTYNAME', 'TOT_POP','NHWA_MALE','NHWA_FEMALE','H_MALE','H_FEMALE']]
WhiteHispanicVPOC_percentile.head()
```

```python
sum_column = WhiteHispanicVPOC_percentile['NHWA_MALE']+ WhiteHispanicVPOC_percentile['NHWA_FEMALE']+ WhiteHispanicVPOC_percentile['H_MALE']+WhiteHispanicVPOC_percentile['H_FEMALE']
WhiteHispanicVPOC_percentile['HISPANIC OR WHITE'] = sum_column

sum_column = WhiteHispanicVPOC_percentile['TOT_POP'] - WhiteHispanicVPOC_percentile['HISPANIC OR WHITE']
WhiteHispanicVPOC_percentile['POC'] = sum_column

WhiteHispanicVPOC_percentile = WhiteHispanicVPOC_percentile[['STNAME','CTYNAME', 'TOT_POP', 'HISPANIC OR WHITE','POC']]
WhiteHispanicVPOC_percentile.head()
```

```python
sum_column = WhiteHispanicVPOC_total['NHWA_MALE']+ WhiteHispanicVPOC_total['NHWA_FEMALE']+ WhiteHispanicVPOC_total['H_MALE']+WhiteHispanicVPOC_total['H_FEMALE']
WhiteHispanicVPOC_total['HISPANIC OR WHITE'] = sum_column

sum_column = WhiteHispanicVPOC_total['TOT_POP'] - WhiteHispanicVPOC_total['HISPANIC OR WHITE']
WhiteHispanicVPOC_total['POC'] = sum_column

WhiteHispanicVPOC_total = WhiteHispanicVPOC_total[['STNAME','CTYNAME', 'TOT_POP', 'HISPANIC OR WHITE','POC']]
WhiteHispanicVPOC_total.head()
```
