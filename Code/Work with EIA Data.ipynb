```python
census = pd.read_csv("cc-est2019-alldata.csv", encoding= 'latin-1')
```

```python
len(census.CTYNAME.unique())
```

```python
census = census[census.AGEGRP == 0]
```

```python
census = census[census.YEAR == 12]
```

```python
census.head()
```

```python
census = census.drop(columns=['SUMLEV', 'STATE','COUNTY','YEAR','AGEGRP'])
census.head()
```

```python
percentile = census.iloc[:,0:2]
percentile.head()
```

```python
population = census.iloc[:,2:]
population.head()
```

```python
population = population.astype(float)
```

```python
population = population.iloc[:,:].div(population.TOT_POP, axis=0)*100
population.head()
```

```python
percentile = percentile.join(population)
percentile.head()
```
