```python
WHITEVPOC_PERC = pd.read_csv("WHITEVPOC_percentile.csv")
WHITEHISPANICVPOC_PERC = pd.read_csv("WHITEHISPANICVPOC_percentile.csv")
WHITEHISPANICVPOC = pd.read_csv("WHITEHISPANICVPOC_total.csv")
WHITEVBLACKVHISPANIC_PERC = pd.read_csv("WHITEVBLACKVHISPANIC_percentile.csv")
WHITEVBLACKVHISPANIC = pd.read_csv("WHITEVBLACKVHISPANIC_total.csv")
```

```python
WHITEVPOC_PERC= WHITEVPOC_PERC.rename(columns={"WA": "WA_perc", "POC": "POC_perc"})
```

```python
WHITEVPOC_PERC = WHITEVPOC_PERC[['STNAME','CTYNAME','WA_perc','POC_perc']]
```

```python
WHITEVPOC = WHITEVPOC.merge(WHITEVPOC_PERC)
```

```python
WHITEHISPANICVPOC_PERC.head()
```

```python
WHITEHISPANICVPOC_PERC= WHITEHISPANICVPOC_PERC.rename(columns={"HISPANIC OR WHITE": "HISPANIC OR WHITE_perc", "POC": "POC_perc"})
```

```python
WHITEHISPANICVPOC_PERC = WHITEHISPANICVPOC_PERC[['STNAME','CTYNAME','HISPANIC OR WHITE_perc','POC_perc']]
```

```python
WHITEHISPANICVPOC = WHITEHISPANICVPOC.merge(WHITEHISPANICVPOC_PERC)
```

```python
WHITEVBLACKVHISPANIC_PERC.head()
```

```python
WHITEVBLACKVHISPANIC_PERC = WHITEVBLACKVHISPANIC_PERC[['STNAME','CTYNAME','NHWA_perc','NHBA_perc','H_perc']]
```

```python
WHITEVBLACKVHISPANIC = WHITEVBLACKVHISPANIC.merge(WHITEVBLACKVHISPANIC_PERC)
```
